1) iris데이터를 이용한 클러스터 예제에서 실루엣 계수를 계산하는 라인을 추가하여 Petal width와 petal length를 사용했을 때 클러스터를 평가해보자

```python
# code for problem 1
from sklearn import cluster,datasets
from sklearn import metrics

import matplotlib.pyplot as plt

iris=datasets.load_iris()
X=iris.data[:,2:]
y=iris.target

print('실제 dataset target별 분류:')
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis')
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
k_means=cluster.KMeans(n_clusters=3).fit(X)
y_pred=k_means.predict(X)

S=metrics.silhouette_score(X,y_pred)
print("silhouette_score:{:1f}\n".format(S))


plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='viridis',edgecolors='k')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1]
           ,c=[0,1,2],cmap='viridis',marker='^',s=200,edgecolors='k')
plt.axis([0,X[:,0].max()+0.2,0,X[:,1].max()+0.2])

print('k-means clustering결과:')
plt.show()
```

### problem1의 결과
실루엣 계수는 0.660480이였고, 데이터셋에 존재하는 실제 결과와 비교해봤을 때 3가지의 꽃의 종류를 잘 분리해냈다.

2) Sepal width와 sepal length를 사용하여 동일한 클러스터 예제를 수행해보고, 시각화 및 실루엣 계수를 계산하여 1)의 결과와 비교해보자

```python
# code for problem 2
from sklearn import cluster,datasets
from sklearn import metrics

import matplotlib.pyplot as plt

iris=datasets.load_iris()
X=iris.data[:,:2]
y=iris.target

print('실제 dataset target별 분류:')
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis')
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
k_means=cluster.KMeans(n_clusters=3).fit(X)
y_pred=k_means.predict(X)
S=metrics.silhouette_score(X,y_pred)
print("silhouette_score:{:1f}\n".format(S))

plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='viridis',edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1]
           ,c=[0,1,2],cmap='viridis',marker='^',s=200,edgecolors='k')
plt.axis([0,X[:,0].max()+0.2,0,X[:,1].max()+0.2])

print('k-means clustering결과:')
plt.show()
```

### problem2의 결과
실루엣 계수가 1)의 경우보다 작게 나왔고, 실제 dataset과 clustering결과를 비교해봤을 때 1)의 경우보다 3가지 종류를 정확하게 구분했다고는 말하기 어렵다.
<br/><br/>
즉, sepal lenth 와 sepal width를 사용하였을 때 보다 petal length와 petal width를 사용하였을 때 더 잘 clustering되었다고 할 수 있다.
