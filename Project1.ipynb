1) iris데이터를 이용한 클러스터 예제에서 실루엣 계수를 계산하는 라인을 추가하여 Petal width와 petal length를 사용했을 때 클러스터를 평가해보자

```python
# code for problem 1
from sklearn import cluster,datasets
from sklearn import metrics

import matplotlib.pyplot as plt

iris=datasets.load_iris()
X=iris.data[:,2:]
y=iris.target

n_clusters=[2,3,4,5,6,7,8]

for k in n_clusters:
    k_means=cluster.KMeans(n_clusters=k).fit(X)
    y_pred=k_means.predict(X)
    S=metrics.silhouette_score(X,y_pred)
    print("n_clusters:{:d},silhouette_score:{:1f}".format(k,S))
```

```python
from sklearn import metrics
from sklearn.cluster import KMeans
k_means=cluster.KMeans(n_clusters=3).fit(X)
y_pred=k_means.predict(X)
S=metrics.silhouette_score(X,y_pred)
print("silhouette_score:{:1f}".format(S))


plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='viridis',edgecolors='k')
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1]
           ,c=[0,1,2],cmap='viridis',marker='^',s=200,edgecolors='k')
plt.axis([0,X[:,0].max()+0.2,0,X[:,1].max()+0.2])

plt.show()
```
